```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import holoviews as hv
hv.extension('bokeh')
```

```python
df = pd.read_csv("../data/process_changepoint.csv")
# lsl, usl = 95, 120
```

```python
data = df.iloc[:, 0].tolist()
plt.ylim([90, 130])
plt.scatter(range(len(data)), data, s=5)
```

Something went wrong at around 60? But what exactly?

```python
from sklearn.ensemble import IsolationForest

data2d = np.array(data).reshape(-1, 1)
clf = IsolationForest(random_state=0)
clf.fit(data2d)
pred = clf.predict(data2d)

c = ['r' if x < 0 else 'b' for x in pred]
plt.ylim([90, 130])
plt.scatter(range(len(data)), data, s=5, c=c)
```

```python
window = 10
moving_average = df.rolling(window).mean().fillna(method='bfill')
plt.ylim([90, 130])
plt.scatter(range(len(moving_average)), moving_average, s=5)
```

```python
moving_average2d = np.array(moving_average).reshape(-1, 1)
clf = IsolationForest(random_state=0)
clf.fit(moving_average2d)
pred = clf.predict(moving_average2d)
```

```python
c = ['r' if x < 0 else 'b' for x in pred]
plt.ylim([90, 130])
plt.scatter(range(len(moving_average)), moving_average, s=5, c=c)
```
